```python
import numpy as np
from keras.applications import resnet
from keras.utils import load_img, img_to_array
from os import listdir
from os.path import isfile, join
```

### I. Loading model

```python
model = resnet.ResNet50()
```

### II. Reading images

#### 1. Images paths

```python
path = 'images/'
file_paths = [join(path, f) for f in listdir(path) if isfile(join(path, f))]
```

#### 2. Reading images

```python
photos = [load_img(f, target_size=(224, 224)) for f in file_paths]
```

```python
photos[0]
```

#### 3. Convert images to arrays

```python
images = [img_to_array(image) for image in photos]
```

#### 4. Scale the input images to the range used in the neural network

```python
images = [resnet.preprocess_input(image) for image in images]
```

### III. Predictions

```python
pred = model.predict(np.array(images)) 
```

```python
pred_classes = resnet.decode_predictions(pred)
```
